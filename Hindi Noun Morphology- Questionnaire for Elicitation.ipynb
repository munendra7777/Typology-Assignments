#### 3. Clitic particles
There are several clitics in Hindi which perform a range of pragmatically and semantically restrictive functions. <br>

- Clitic *'hii'*: it adds exclusive focus(EXCL_FOC) to its arguement depending upon subject or object of focus.<br>

For example: 


|Arijit ne hii|gaane|gaaye|
|:------------:|-------|--------------|
|*Arjit-N ne-ERG hii-EXCL_FOC*| *gaane-Songs*| *gaaye-Sing.Perf.M.PL*|
|It was (only) Arijit who sang songs |

 |Arijit ne |gaane hii|gaaye|
 |:------------:|-------|--------------|
 |*Arjit-N ne-ERG*| *gaane-Song hii-EXCL_FOC*| *gaaye-Sing.Perf.M.PL*|
 |It was (only) songs which Arijit sang|

- clitic *bhii*: it adds inclusive focus (INCL_FOC) to its arguement depending upon subject or object of focus.<br>
For Example:

|Arijit ne bhii|gaane|gaaye|
|:------------:|-------|--------------|
|*Arjit-N ne-ERG bhii-INCL_FOC*| *gaane-Songs*| *gaaye-Sing.Perf.M.PL*|
|Arijit (also) sang songs |

 |Arijit ne|gaane bhii|gaaye|
 |:------------:|-------|--------------|
 |*Arjit-N ne-ERG*| *gaane-Song bhii-INCL_FOC*| *gaaye-Sing.Perf.M.PL*|
 |Arijit sang songs(also) |
 
 - Clitic *to*: it adds contrastive topic (CONT_TOP) to its arguement depending upon subject or object of focus.<br>
For Example:


  |Cigarette to | mili |lekin ab maachis|kho gyi|
  |:------|------|:-------|:-------|
  |*Cigarette-N.SG.CONT_TOP* | *found-PERF.F.SG* | *but now matches-N* | *lost go.PERF.F.PL* |
  |The cigarette was found but 
  now the matches are lost. |
  
  |Cigarette|mili|to ab maachis|kho gyi|
  |:-------|:------|:-------|:---------|
  |*Cigarette-N.SG* | *found-PERF.F.SG*| *CONT_TOP.now matches-N.PL* |*lost go.PERF.F.PL*|
  |The cigarette was found but
  now the matches are lost.|
  
<br>
  
   (*See: Devyani Sharma 1999,Nominal Clitics and Constructive Morphology in Hindi*)

#### 7. Object of Equation 
For example: He is as big as me.

Hindi in equative clauses uses an equative adjective or adverb  with the

Equative sentences can also be formed by using the clause '*jaisa: /jaise* (as/which way)' and '*vaisa:/vaise* (like/that way). <br>

#### 14. Topic of a sentence
The "topic sentence" is the sentence in which the main idea of the paragraph is stated. It is unquestionably the most important sentence in the paragraph. The topic sentence generally is composed of two parts: (a) the topic itself and (b) the controlling idea.

|Yog ke |anek fa:yade hain|
|--------------|-----------|
|Yoga.GEN|many benefits are|
|*topic* |*controlling idea*|
|Yoga has many benefits.|
