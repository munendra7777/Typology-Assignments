# What is *Agreement* in a sentence?

Agreement is the correspondence of a verb with its subject in person and number, and of a pronoun with its antecedent in person, number, and gender. Another term for grammatical agreement is concord.

In Hindi, there is agreement in terms of *number, person and gender.*

|------|------|------|
|Singular|First|Masculine|
|Plural|Second|Feminine|

# What is the *Subject* in a sentence?

The subject is a grammatical term used to describe the nouns, pronouns, and noun phrases that occur before the verb in a sentence. Thus, the subject describes a position in the sentence. When the verb is in the active form, the subject of the sentence is the *do-er* or agent—who or what—that causes the action.

## What is Ergativity?

Ergativity is an grammatical pattern in which the subject of an intransitive clause is treated in the same way as the object of a transitive clause, and differently from transitive subject. 
*(Dixon 1994: 1)*

Hindi and Awadhi both are split-ergative language which means that the ergative alignment is only found in particular constructions and/or paradigms. In Hindi and Awadhi, the ergative pattern appears in the preterit and perfect tenses. Ergativity is formally marked by the clitic *ne*. 
*(Coon, 2013)*

A common type of split ergativity is TAM split ergativity, which means that the split in alignment is conditioned by the tense, aspect, and/or mood of the verb. *(Dixon, 1994)* This type is found in Hindi and Awadhi, where the case on the subject of a finite, transitive clause depends upon the aspect.

> - For example:

  |Subject|Predicate|language|
  |-------|----------|-------|
  |Arijit ne|kai: ga:ne ga:ye|Hindi|
  |Arijit.ERG|many songs sing.V.PERF|
  |Arijitwa: ne|kaiu ga:nan ga:yen|Awadhi|
  |Arijit.ERG|many songs sing.V.PERF|
  |Arijit|sang many songs|English|

*Note:The verb **ga:ye/ga:yen** is marked for perfective aspect **-ye/-yen**, and the subject **Arijit/Arijitwa:** bears the ergative case marker **-ne** respectively in Hindi and Awadhi.*

  |Subject|Predicate|Language|
  |-------|----------|-------|
  |Arijit |ga:ne ga:ta: hai/tha:|Hindi|
  |Arijit| songs sing.V.HAB.be.PRES/.PAST|
  |Arijitwa:|ga:nan (ko) ga:wat hai:/rahai:|Awadhi|
  |Arijit| songs sing.V.HAB.be.PRES/.PAST|
  |Arijit|sings/ used to sing songs|


*Note: When the lexical verb is not marked for perfective aspect as in verb above **ga:ta:/ga:wat** is a habitual form of the verb **ga:na:(to sing)**, the subject Arijit does not receive ergative case. Tense has no effect on the case for the subject in Hindi and Awadhi and the subject Arijit remains unmarked for case.*

Hindi and Awadhi allows some verbs for optional ergative case marking on their subject—i.e., both nominative and ergative
subjects are grammatical. This holds true for transitive as well as intransitive verbs.

> - For example:

  |Subject|Predicate|Language|
  |-------|---------|--------|
  |Ra:m /ne|zorse kha:sa:|Hindi|
  |Ram/.ERG|loudly cough.PRF|
  |Ramwa:/ne|zorse kha:so:|Awadhi|
  |Ram/.ERG|loudly cough.PRF|
  |Ram|coughed loudly.|English|

  |Subject|Predicate|Language|
  |-------|---------|--------|
  |Ra:m /ne|socha: ye kita:b meri: hai|Hindi|
  |Ram/.ERG|think.PERF this.DET book.N I.GEN is|
  |Ramwa /ne|socho: i: kita:b hama:ri: hai:|Awadhi|
  |Ram/.ERG|think.PERF this.DET book.N I.GEN is|
  |Ram|thought this book was mine|English|
 
- Some transitive verbs normally takes *NOM* but *ERG* is possible in Hindi.

|Verbs|Case|
|---------|---------|
|bhuul(forget)|NOM/ERG|
|jan(give birth to)|NOM/ERG|
|phañ:d (leap over)|NOM/ERG|
|bak (to say nonsense)|NOM/ERG|
|ha:r(lose[be defeated])NOM|(lose on purpose) ERG|
|pa: (INF-manage,succeed)|NOM/ERG|
|qara:r pa: (obtain rest) |NOM/ERG|
|pahcha:n (recognize)|NOM/ERG|

Although Awadhi slightly varies phonologically and morphologically from Hindi but it closely follows Hindi.
